**For numerical data** 

**Use the mean and the standard deviation as measures of center distributions that are reasonably symmetric with a central peak.**

**When distribution is skewed, the mean and standard deviation are not a good choice. Use median, IQR for Skewed distribution**


**Here Finding Skewness of columns and **
if skewness is between -5 < skew() < 5  - symmetrical distribution**

**if skewness is between -1 <skew() -5 and 5   - modular skewed  distribution**

**if skewness is between skew() < -1 and  1 < skew() - symmetrical distribution**

**So measure of central tendency i.e mean and measure of dispersion i.e standard devision(sd) for symmetrical distribution**

**So measure of central tendency i.e median and measure of dispersion i.e IQR for skewed distribution.**

##  Checking Age Column in dataset is skewed or symmetric ?

```python
sns.histplot(data = olympix['age'], kde = True)   # plot histograph
plt.axvline(x=olympix.age.mean(), color = 'red', alpha = 0.5, label = 'Mean')  # draw mean value in plot
plt.axvline(x=olympix.age.median(), c = 'blue', ls = '--', alpha = 0.5, label = 'Mean')   # draw median value in plot
plt.legend
```

Age plot is Right Skewed

```python
olympix['age'].skew()     # calculating skewedness of age
```

value of skewness > 5 so perform Transformation to get skewness between -5 and 5

#### Log Transformation

```python
olympix['age_log'] = np.log(olympix['age'])  # log Transformation
```

```python
sns.histplot(data = olympix['age_log'], kde = True) # plot histograph
plt.axvline(x=olympix.age_log.mean(), color = 'red', alpha = 0.5, label = 'Mean')  # # draw mean value in plot
plt.axvline(x=olympix.age_log.median(), c = 'blue', ls = '--', alpha = 0.5, label = 'Mean')  # draw median value in plot
plt.legend
```

get symmetrical age plot

```python
olympix['age_log'].skew()
```

get skewness between -5 and 5  So age plot has symmetrical distribution

####   Count plot of Variables  gold medal, silver medal, brone medal, total medal

```python
n=1
plt.subplots(figsize=(20, 16))  # ploting subplots
for i in col:                  # for that use for loop
    plt.subplot(3,2,n)     # subplot dimensions 
    ax = plt.gca()         # axis view
    sns.set(font_scale = 1.5)     # scale
    sns.countplot(x = olympix[i])     # plot count plot
    ax.set_title('Distribution of %s'%(i), fontsize=23)  # give tile 
    n+=1 
    plt.tight_layout()
plt.suptitle('Distribution of variables', y=1.03, fontsize=30) # give title to whole plot
plt.show() # show plot
```
